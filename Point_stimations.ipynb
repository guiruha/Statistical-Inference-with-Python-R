## ESTIMACION PUNTUALES EN PYTHON

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt
```

VAMOS A ELEGIR UNA m.a.s. de TAMAÑO 200 de una variable aleatoria de
$N(0, 1)$

```python
sample = stats.norm.rvs(loc = 0, scale = 1,  size = 200)
```

```python
sample[:10]
```

Para estimar los parámetros usamos la distribución que consideramos adecuada (en este caso la normal) e invocamos la función 'fit'

```python
mu, std = stats.norm.fit(sample)
```

```python
'la media es igual a {:.4} desviación típica igual a {:.4}'.format(mu, std)
```

A continuación compraremos los parámetros estimados con nuestra muesta

```python
x = np.linspace(-5, 5, 100)
pdf_fitted = stats.norm.pdf(x, loc = mu, scale = std)
pdf_original = stats.norm.pdf(x, loc = 0, scale = 1)
```

```python
%matplotlib inline
fig= plt.subplots(1, 1, figsize = (15, 8))
plt.plot(x, pdf_fitted, 'r-*')
plt.plot(x, pdf_original, 'b--')
plt.hist(sample, color = 'g', normed = True, alpha = 0.5)
```

## EJERCICIO DISTRIBUCIÓN DE RAYLEIGH

```python
sample = stats.rayleigh.rvs(loc=2, scale=2, size = 150)
mean, desv = stats.rayleigh.fit(sample)
print("Media = {} y Desviación Típica = {}".format(mean, desv))
```
